```python
import math
import pandas as pd

feh_hero_dict = {}
feh_weapons_dict = {}
feh_assits_dict = {}
feh_specials_dict = {}
feh_passives_dict = {}
for i, name in enumerate(names_fixed):
    response = requests.get('https://feheroes.wiki/{0}'.format(name))
    content = response.content
    parser = BeautifulSoup(content, 'html.parser')
    skills_table = parser.select('.skills-table')
    table_heads = []
    table_data = []
    dataframes = []
    
    for j in skills_table:
        headers = j.find_all('th')
        temp_hlist = []
        for k in headers:
            temp_hlist.append(k.text)
            
        data = j.find_all('td')
        temp_dlist = []
        for l in data:
            temp_dlist.append(l.text)
            
        table_heads.append(temp_hlist)
        table_data.append(temp_dlist)
        
    for index, each in enumerate(table_heads):
        if each[1] == 'Effect':
            intended_passive_total = math.ceil(len(table_data[index])/len(table_heads[index]))
            ipt = intended_passive_total
            #print(ipt)
            passive_fixed = table_data[index][:]
            for m in range(0, ipt):
                start = m * len(table_heads[index])
                end = ((m+1) * len(table_heads[index])) - 1
                #print('end: ', end)
                if m < 3:
                    current_type = table_data[index][4]
                elif m < 6 and m > 2:
                    current_type = table_data[index][17]
                elif m < 9 and m > 5:
                    current_type = table_data[index][30]
                #print(current_type)
                if end > len(passive_fixed) - 1:
                    passive_fixed.insert(end, current_type)
                elif passive_fixed[end] != current_type:
                    passive_fixed.insert(end, current_type)
            table_data[index] = passive_fixed
            #print(table_data[index])
            #print(passive_fixed)
    
    for index, item in enumerate(table_heads):
        #print(len(item), len(table_data[index]))
        #print(table_data[index])
        rows = []
        for i in range(0, int(len(table_data[index])/len(item))):
            temp_l = []
            start = i * len(item)
            for j in range(0, len(item)):
                current = start + j
                temp_l.append(table_data[index][current])
            rows.append(temp_l)
        #print(rows)
        df = pd.DataFrame(rows, columns=item)
        dataframes.append(df)
    #print(dataframes)
    feh_hero_dict[name] = dataframes
                
    

        
#print(feh_hero_dict.keys())
#print(feh_hero_dict.values())
#print(feh_hero_dict['Abel'][0])

for key, value in feh_hero_dict.items():
    print(key)
    print(value)
```

```python
def sorted_dataframes(feh_dict):
    weapons = []
    assists = []
    specials = []
    passives = []
    count = 0
    for key, value in feh_dict.items():
        for each in value:
            if each.columns[1] == 'Might':
                #print(each.shape[0])
                #print(each)
                key_list = []
                for i in range(0, each.shape[0]):
                    key_list.append(key)
                #print(key_list)
                new_add = each.copy()
                new_add['Hero'] = key_list
                weapons.append(new_add)
            elif each.columns[1] == 'Range':
                key_list = []
                for i in range(0, each.shape[0]):
                    key_list.append(key)
                new_add = each.copy()
                new_add['Hero'] = key_list
                assists.append(new_add)
            elif each.columns[1] == 'Cooldown':
                key_list = []
                for i in range(0, each.shape[0]):
                    key_list.append(key)
                new_add = each.copy()
                new_add['Hero'] = key_list
                specials.append(new_add)
            elif each.columns[1] == 'Effect':
                key_list = []
                for i in range(0, each.shape[0]):
                    key_list.append(key)
                new_add = each.copy()
                new_add['Hero'] = key_list
                passives.append(new_add)
        #print(weapons)
        character_info = []
        if len(weapons) > 0:
            character_info.append(weapons)
        if len(assists) > 0:
            character_info.append(assists)
        if len(specials) > 0:
            character_info.append(specials)
        if len(passives) > 0:
            character_info.append(passives)
        
        #count += 1
        #if count > 0:
        #    break
            
    return(character_info)

                
#add each dataframe to the respective list, also add a new column and fill the column with the name of hero it
#belongs to. one way to do this is use DataFrame.shape[0] with gives you how many rows are in the dataframe.
#make a list that length, and fill it with the key for that dataframe. then DataFrame[key][each]['Hero'] = list
#that will create a column named hero, which all dataframes will have, and add the list as an unique ID


sorted_feh = sorted_dataframes(feh_hero_dict)
print(sorted_feh)
```
